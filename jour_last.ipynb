###### נבצע מספר שינויים בדאטה פריים: הוספת קטגוריות ושינויי type

```python
jour['Date'] = pd.to_datetime(jour['Date'], errors='coerce')
jour['Year'] = jour['Date'].dt.year
jour['Month'] = jour['Date'].dt.month
```

```python
# separate by comma to make SubCategories

jour['Coverage'] = jour['Coverage'].fillna('').astype(str)
```

```python
jour['Classify cover'] = jour['Coverage'].apply(lambda x: x.split(', '))
```

```python
#  הפרדה של עמודת הסיקור לקטגוריות מוגדרות ללא פסיקים

unique_coverage = set([category for sublist in jour['Classify cover'] for category in sublist])
print(unique_coverage)
```

```python
for category in unique_coverage:
    jour[category] = jour['Classify cover'].apply(lambda x: category in x)
```

```python
###### כאן, נהפוך את העמודות שיצרנו (שהיו בוליאנים) לטייפ אינט כדי לסכום בקלות בהמשך במידה ונצטרך

jour[['War', 'Sports', 'Human Rights', 'Politics', 'Culture','Business', 'Corruption', 'Crime']] = jour[['War', 'Sports', 'Human Rights', 'Politics', 'Culture','Business', 'Corruption', 'Crime']].astype(int)
```

```python
# החלפת כל המילים הרלוונטיות ל-Freelancer בעזרת loc
jour.loc[jour['Organization'].isin(['Freelance', 'freelance', 'Freelancer']), 'Organization'] = 'Freelancer'
```

```python
# חישוב עשרת הארגונים החוזרים על עצמם הכי הרבה
top_10_organizations = jour['Organization'].value_counts().head(10)

# חישוב הלאום החוזר על עצמו הכי הרבה
top_10_Nationality = jour['Nationality'].value_counts().head(10)

# חישוב העבודה שחוזרת על עצמה הכי הרבה
top_10_Job = jour['Job'].value_counts().head(10)

# חישוב המדינות שחוזרות הכי הרבה
top_10_Country = jour['Country_killed'].value_counts().head(10)

# יצירת גרף 2x2 עם Seaborn ו-Matplotlib
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# גרף 1 - ארגונים
sns.barplot(ax=axes[0, 0], x=top_10_organizations, y=top_10_organizations.index, palette='viridis')
axes[0, 0].set_title('Top 10 Organizations')
axes[0, 0].set_xlabel('Journalists killed')
axes[0, 0].set_ylabel('Organization')

# גרף 2 - לאום
sns.barplot(ax=axes[0, 1], x=top_10_Nationality, y=top_10_Nationality.index, palette='viridis')
axes[0, 1].set_title('Top 10 Journalists Nationality')
axes[0, 1].set_xlabel('Journalists killed')
axes[0, 1].set_ylabel('Nationality')

# גרף 3 - תפקיד
sns.barplot(ax=axes[1, 0], x=top_10_Job, y=top_10_Job.index, palette='viridis')
axes[1, 0].set_title('Top 10 Journalists Job')
axes[1, 0].set_xlabel('Journalists killed')
axes[1, 0].set_ylabel('Job')

# גרף 4 - מדינות
sns.barplot(ax=axes[1, 1], x=top_10_Country, y=top_10_Country.index, palette='viridis')
axes[1, 1].set_title('Top 10 Journalist Deaths in These Countries')
axes[1, 1].set_xlabel('Journalists killed')
axes[1, 1].set_ylabel('Country')

# התאמת עיצוב הגרפים
plt.tight_layout()
plt.show()
```
